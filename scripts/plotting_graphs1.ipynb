```python
df_n = pd.read_csv('data/insights/title_desc_cap/hi_ndtv_india_insights_tdc.csv')
```

```python
# Calculate the counts of each target party
target_party_counts = df_n['Target Party'].value_counts()

# Define a threshold for grouping small proportions into "Other"
threshold = 0.02 * target_party_counts.sum()  # 2% threshold
target_party_counts = target_party_counts[target_party_counts >= threshold]
other_count = df_n['Target Party'].value_counts()[df_n['Target Party'].value_counts() < threshold].sum()

# Add the "Other" category if needed
if other_count > 0:
    target_party_counts['Other'] = other_count

# Calculate the proportions
target_party_proportions = target_party_counts / target_party_counts.sum()

# Plot the pie chart
colors = ['#f97d09','#02478E','#167B37','#B02C28']
plt.figure(figsize=(10, 7), facecolor='none')  
plt.pie(
    target_party_proportions,
    labels=target_party_proportions.index,
    autopct='%1.1f%%',
    startangle=140,
    textprops={'color': 'white'} ,
    colors=colors[:len(target_party_proportions)] 
)
plt.title('Party Mentions in the Videos', color='white')  
plt.gca().set_facecolor('none')
```

```python
# Ensure the 'publish_date' column is in datetime format
# df_n.loc[:, 'publish_date'] = pd.to_datetime(df_n['publish_date'], format='%d-%m-%Y')
# Convert 'publish_date' to datetime format with automatic inference for mixed formats
df_n['publish_date'] = pd.to_datetime(df_n['publish_date'], format='mixed', dayfirst=True)


# Define custom colors for the parties
party_colors = {
    'BJP': '#f97d09',
    'INC': '#02478E',
    'AAP': '#167B37', 
}

# Function to plot the timeline for a given orientation
def plot_timeline(orientation, title, ax, threshold, party_colors):
    # Filter for rows where Orientation matches the specified orientation
    df_orientation = df_n[df_n['Orientation'] == orientation]

    # Define the threshold and filter out parties below the threshold
    party_counts = df_orientation['Target Party'].value_counts()
    filtered_parties = party_counts[party_counts >= threshold].index
    df_orientation = df_orientation[df_orientation['Target Party'].isin(filtered_parties)]

    # Group the data by publish_date and Target Party to get the count of sentiments
    grouped_df = df_orientation.groupby(['publish_date', 'Target Party']).size().unstack(level=1, fill_value=0)
    
    # Plot the timeline for each party with custom colors
    colors = [party_colors.get(col, '#333333') for col in grouped_df.columns]
    grouped_df.plot(ax=ax, linestyle='-', color=colors)

    # Set labels with white color
    ax.set_xlabel('Date', color='white')
    ax.set_ylabel('Count', color='white')
    ax.set_title(title, color='white')

    # Set x and y axis tick parameters to white
    ax.tick_params(axis='x', colors='white')
    ax.tick_params(axis='y', colors='white')

    # Set the figure background and axes face color to transparent
    fig.patch.set_alpha(0.0)
    ax.patch.set_alpha(0.0)
    
    # Set the border color of the plot to white
    ax.spines['top'].set_color('white')
    ax.spines['right'].set_color('white')
    ax.spines['bottom'].set_color('white')
    ax.spines['left'].set_color('white')

# Define thresholds
threshold_supports = 10
threshold_opposes = 30

# Create a plot for 'Supports' and 'Opposes' sentiments side by side
fig, axes = plt.subplots(1, 2, figsize=(15, 7), sharey=True)

# Plot for 'Supports' sentiment
plot_timeline('Supports', 'Timeline of Supportive Sentiments by Party', axes[0], threshold_supports, party_colors)

# Plot for 'Opposes' sentiment
plot_timeline('Opposes', 'Timeline of Opposing Sentiments by Party', axes[1], threshold_opposes, party_colors)


plt.tight_layout()
plt.show()
```

```python
# Drop rows where 'publish_date', 'Orientation' or 'Target Party' is NaN
df = df_n.dropna(subset=['publish_date', 'Orientation', 'Target Party'])

# Convert 'publish_date' to datetime
df['publish_date'] = pd.to_datetime(df['publish_date'])

# Define the phases and their dates
phases = {
    'Phase 01': '2024-04-19',
    'Phase 02': '2024-04-26',
    'Phase 03': '2024-05-07',
    'Phase 04': '2024-05-13',
    'Phase 05': '2024-05-20',
    'Phase 06': '2024-05-25',
    'Phase 07': '2024-06-01'
}

# Add a phase column to the dataframe
def assign_phase(date):
    for phase, phase_date in phases.items():
        if date <= pd.to_datetime(phase_date):
            return phase
    return 'Phase 07'

df['Phase'] = df['publish_date'].apply(assign_phase)

# Group by 'Phase' and 'Target Party' and count the occurrences for each orientation
supports_counts = df[df['Orientation'] == 'Supports'].groupby(['Phase', 'Target Party']).size().unstack(fill_value=0)
opposes_counts = df[df['Orientation'] == 'Opposes'].groupby(['Phase', 'Target Party']).size().unstack(fill_value=0)

# Define a threshold to show only the top parties
threshold = 10


# Filter the parties based on the threshold
top_supports_parties = supports_counts.columns[supports_counts.sum() > threshold]
top_opposes_parties = opposes_counts.columns[opposes_counts.sum() > threshold]

supports_counts = supports_counts[top_supports_parties]
opposes_counts = opposes_counts[top_opposes_parties]

# Define custom colors for the parties
party_colors = {
    'BJP': '#f97d09',
    'INC': '#02478E',
    'AAP': '#167B37', 
}

# Plot the horizontal bar charts side by side
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 10), sharey=True)

# Set transparent background
fig.patch.set_alpha(0)
for ax in axes:
    ax.patch.set_alpha(0)

# Plot the supports chart
supports_counts.plot(kind='barh', stacked=False, ax=axes[0], color=[party_colors.get(col, '#333333') for col in supports_counts.columns])
axes[0].set_title('Supports', color='white')
axes[0].set_xlabel('Counts', color='white')
axes[0].set_ylabel('Phase', color='white')
axes[0].legend(title='Target Party', facecolor='white', edgecolor='white', labelcolor='black')

# Plot the opposes chart
opposes_counts.plot(kind='barh', stacked=False, ax=axes[1], color=[party_colors.get(col, '#333333') for col in opposes_counts.columns])
axes[1].set_title('Opposes', color='white')
axes[1].set_xlabel('Counts', color='white')
axes[1].set_ylabel('', color='white')

# Set white text and border for both axes
for ax in axes:
    ax.title.set_color('white')
    ax.xaxis.label.set_color('white')
    ax.yaxis.label.set_color('white')
    ax.spines['bottom'].set_color('white')
    ax.spines['top'].set_color('white')
    ax.spines['right'].set_color('white')
    ax.spines['left'].set_color('white')
    ax.tick_params(axis='x', colors='white')
    ax.tick_params(axis='y', colors='white')
    for label in ax.get_yticklabels():
        label.set_color('white')
    for label in ax.get_xticklabels():
        label.set_color('white')

plt.suptitle('Party-wise Sentiment Counts Across Election Phases', color='white')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

```
