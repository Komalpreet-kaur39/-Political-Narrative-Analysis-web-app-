# **Data Cleaning**

## **NDTV**

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/en_top3_insights_NDTV.csv')
df.head(5)
```

```python
df.isna().sum()
```

```python
df['Target Party'].unique()
```

```python
df['Target Party'].isna().sum()
```

```python
df['Target Party'] = df['Target Party'].replace('BJP(Bharatiya Janta Party)', 'BJP')
df['Target Party'] = df['Target Party'].replace('INC(Indian National Congress)', 'INC')
df['Target Party'] = df['Target Party'].replace('INDIA bloc (which includes INC and AAP)', 'INC')
df['Target Party'] = df['Target Party'].replace('Congress', 'INC')
df['Target Party'] = df['Target Party'].replace('Trinamool Congress', 'TMC')
df['Target Party'] = df['Target Party'].replace('AAP(Aam Aadmi Party)', 'AAP')
df['Target Party'] = df['Target Party'].replace('TMC(Trinamool Congress)', 'TMC')
df['Target Party'] = df['Target Party'].replace('TMC (Trinamool Congress)', 'TMC')
df['Target Party'] = df['Target Party'].replace('BRS (Bharat Rashtra Samithi)', 'BRS')
df['Target Party'] = df['Target Party'].replace('RJD (Rashtriya Janata Dal)', 'RJD')
```

```python
df['Target Party'].unique()
```

```python
df['Target Party'] = df['Target Party'].replace('YSR Congress Party (YSRCP)', 'YSRCP')
df['Target Party'] = df['Target Party'].replace('YSR Congress Party', 'YSRCP')
df['Target Party'] = df['Target Party'].replace('JDS (Janata Dal Secular)', 'JDS')
```

```python
df['Target Party'].unique()
```

```python
df = df[df['Target Party'] != 'Conservative Party']
df['Target Party'].unique()
```

```python
df['Target Party'] = df['Target Party'].replace('JMM (Jharkhand Mukti Morcha)', 'JMM')
df['Target Party'] = df['Target Party'].replace('TDP (Telugu Desam Party)', 'TDP')
df['Target Party'] = df['Target Party'].replace('BJD (Biju Janata Dal)', 'BJD')
```

```python
df['Target Party'].unique()
```

```python
# Splitting the 'Target Party' column and creating new rows
df['Target Party'] = df['Target Party'].str.split(' and ')
df = df.explode('Target Party')
df['Target Party'].unique()
```

```python
df['Target Party'].value_counts()
```

**Associating Alliances**

```python
df['Alliance'] = ''
df.columns
```

```python
df['Target Party'].unique()
```

```python
df.loc[df['Target Party'] == 'BJP', 'Alliance'] = 'NDA'
df.loc[df['Target Party'] == 'TMC', 'Alliance'] = 'INDIA'
df.loc[df['Target Party'] == 'INC', 'Alliance'] = 'INDIA'
df.loc[df['Target Party'] == 'BJD', 'Alliance'] = 'Others'
df.loc[df['Target Party'] == 'RJD', 'Alliance'] = 'INDIA'
df.loc[df['Target Party'] == 'BRS', 'Alliance'] = 'Others'
df.loc[df['Target Party'] == 'AAP', 'Alliance'] = 'INDIA'
df.loc[df['Target Party'] == 'JMM', 'Alliance'] = 'INDIA'
df.loc[df['Target Party'] == 'YSRCP', 'Alliance'] = 'Others'
df.loc[df['Target Party'] == 'TDP', 'Alliance'] = 'Others'
df.loc[df['Target Party'] == 'JDS', 'Alliance'] = 'NDA'
df.loc[df['Target Party'] == 'AIMIM', 'Alliance'] = 'Others'
```

```python
df['Target Party'].isna().sum()
```

```python
df['Alliance'].unique()
```

```python
df['Alliance'] = df['Alliance'].replace('', None)
```

```python
df['Alliance'].isna().sum()
```

```python
df.to_csv('/content/en_top3_insights_NDTV.csv', index=False, quoting=1)
```
